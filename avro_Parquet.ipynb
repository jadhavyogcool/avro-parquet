```python
data = [
    (1, "Amit", 85.5, "CS"),
    (2, "Neha", 90.0, "IT"),
    (3, "Raj", 78.0, "CS"),
    (4, "Priya", 92.5, "EXTC"),
    (5, "Karan", 88.0, "IT")
]

columns = ["id", "name", "marks", "department"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
df.write.format("avro").mode("overwrite").save("/Volumes/workspace/default/my_volume/avro_demo")
```

```python
df.write.format("parquet").mode("overwrite").save("/Volumes/workspace/default/my_volume/parquet_demo")
```
